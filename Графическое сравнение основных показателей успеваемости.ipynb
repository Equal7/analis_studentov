```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
stud_df = pd.ExcelFile('datastud.xlsx')
table = stud_df.parse('Результаты')
table
```

```python
table_zachet = table.copy()
table_zachet
```

```python
table_zachet.dropna(axis=0, subset = ['Зачет'], inplace=True)
table_zachet
```

```python
table_zachet.sort_values(by=['Дисциплина', 'Группа', 'Студент (Код)'])
```

```python
table_zachet['Группа'].unique()
```

```python
table_zachet.dropna(axis=0, subset=['Зачет', 'Оценка по предмету'])
```

```python
table_stud = table.iloc[:, [2,6,8,11]].dropna(axis=0).sort_values(axis=0, by=['Студент (Код)'])
table_stud
sns.set_style('darkgrid')
```

```python
Stud_mean = table_stud.groupby('Студент (Код)').mean()
Stud_mean
```

```python
plt.figure(figsize = (12,9))
plt.hist(Stud_mean['Оценка по предмету'], alpha = 0.8, bins = np.arange(3, 5.1, 0.05), color = 'red');
plt.hist(Stud_mean['Средний балл'], alpha = 0.6, bins = np.arange(3, 5.1, 0.05), color = 'green');
plt.hist(Stud_mean['Оценка студента (опрос)'], alpha = 0.4, bins = np.arange(3, 5.1, 0.05), color = 'blue');
plt.legend(['Оценка по предмету', 'Средний балл', 'Оценка студента (опрос)'])
plt.title('Сравнение оценок');
```
